```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
#define matrix vector multiplication
def matrix_vector_multiply(matrix, vector):
    return [sum(matrix[i][j] * vector[j] for j in range(len(vector))) for i in range(len(matrix))]
```

```python
#define plot vectors
def plot_vectors(matrix, original_vectors):
    transformed_vectors = [matrix_vector_multiply(matrix, vector) for vector in original_vectors]
    
    colors = ['blue', 'red']
    labels = ['u', 'v']
    
    plt.figure()
    for i, (original, transformed) in enumerate(zip(original_vectors, transformed_vectors)):
        plt.arrow(0, 0, original[0], original[1], head_width=0.5, head_length=0.3, fc=colors[i], ec=colors[i], label=f'{labels[i]}')
        plt.arrow(0, 0, transformed[0], transformed[1], head_width=0.5, head_length=0.3, fc=colors[i], ec=colors[i], linestyle='--')
        plt.text(original[0], original[1], f'{labels[i]}', ha='right' if i == 0 else 'left', va='bottom')
        plt.text(transformed[0], transformed[1], f'A{labels[i]}', ha='right' if i == 0 else 'left', va='bottom')
    
    plt.xlim(-25, 7)
    plt.ylim(-6, 21)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# Define matrix A and original vectors u and v
A = [[1, 6],
     [5, 2]]
original_vectors = [[6, -5], [3, -2]]

# Call the function to plot vectors
plot_vectors(A, original_vectors)
```
